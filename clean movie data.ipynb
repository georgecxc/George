```python
import pandas
```

```python
df=pandas.read_csv('MoviesOnStreamingPlatform.csv')
df
```

```python
df=df.drop(['Title','ID','Directors'],axis=1)
```

```python
df
```

```python
df.isnull().sum()
```

```python
df=df.dropna(axis=0,subset=['Genres','Country','Language'])
```

```python
def per(x):
    if not pandas.isna(x):
        x=x[:-1]
        return float(x)
```

```python
df['Rotten Tomatoes(%)']=df['Rotten Tomatoes'].apply(per)
```

```python
df
```

```python
df['Genres']=df['Genres'].str.split(',')
```

```python
df['Country']=df['Country'].str.split(',')
```

```python
df['Language']=df['Language'].str.split(',')
```

```python
df
```

```python
def cou(x):
    if 'United States' in x:
        return 1
    else:
        return 0
```

```python
df['isInUS']=df['Country'].apply(cou)
```

```python
def lan(x):
    if 'English' in x:
        return 1
    else:
        return 0
```

```python
df['isInEng']=df['Language'].apply(lan)
```

```python
def mul(x):
    if len(x)>=2:
        return 1
    else:
        return 0
```

```python
df['isMultiCountry']=df['Country'].apply(mul)
```

```python
df['isMultiLanguage']=df['Language'].apply(mul)
```

```python
df1=df.explode('Genres')
```

```python
df2=df.explode('Country')
```

```python
df3=df.explode('Language')
```

```python
df.to_csv('movieData.csv',index=False)
```

```python
df1.to_csv('movieDataGenres.csv',index=False)
```

```python
df2.to_csv('movieDataCountry.csv',index=False)
```

```python
df3.to_csv('movieDataLanguage.csv',index=False)
```
